## Interpolation via zero-padding

This notebook seeks to demonstrate the effect of zero-padding in the incorrect location in the DFT domain

```python
import numpy as np
from scipy.constants import c
import matplotlib.pyplot as plt
```

```python
#%% Function to return the indeces of a DFT
def FFT_freq(N, fp, f0):
    freq = np.arange(N, dtype=int)
    fidx = freq>=(N/2)
    freq[fidx] -= N
    unwrapped_offset = np.round(N*f0/fp).astype(int)
    wrapped_offset = unwrapped_offset%N
    cycle = np.round((unwrapped_offset - wrapped_offset)/N).astype(int)
    
    freq = np.roll(freq, wrapped_offset) + wrapped_offset + cycle*N
    return freq*fp/N
```

```python
"""
Define chirp signal parameters. The rate is caluclated
from the chirp length and the bandwdith
"""
chirp = {"length": 21e-6,
         "bandwidth": 100e6}
chirp["rate"] = 2*np.pi*chirp["bandwidth"]/chirp["length"]
```

```python
"""
Define the sampling frequency and a fast-time array of samples
"""
fs = 120e6 #120 MHz
t = np.arange(-1.5e-5, 1.5e-5, 1/fs)
```

```python
signal_baseband = np.exp(1j*chirp["rate"]/2*t**2)
signal_baseband[abs(t) > chirp["length"]/2] = 0
```

```python
plt.figure()
plt.plot(t, np.unwrap(np.angle(signal_baseband)))
plt.xlabel('time (s)')
plt.ylabel('Pulse phase (rad)')
plt.grid()
plt.show()
```

```python
"""
Mix the signal up to the carrier
"""
fc = 9.5e9 #9.5 GHz carrier
signal_carrier = np.exp(1j*2*np.pi*fc*t)*signal_baseband
signal_carrier[abs(t) > chirp["length"]/2] = 0
```

```python
plt.figure()
plt.plot(t, np.abs(signal_carrier))
plt.xlabel('time (s)')
plt.ylabel('Pulse Response')
plt.grid()
plt.show()
```

```python
N = len(signal_carrier)
signal_carrier_DFT = np.fft.fft(signal_carrier)
```

```python
plt.figure()
plt.plot(range(N), np.abs(signal_carrier_DFT))
plt.xlabel('Sample number of DFT')
plt.ylabel('Response')
plt.grid()
plt.show()
```

```python
f = FFT_freq(N, fs, fc)
```

```python
plt.figure()
plt.plot(range(len(f)), f, '.')
plt.xlabel('Sample number')
plt.ylabel('Associated Frequency (Hz)')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(f,np.abs(signal_carrier_DFT),'.')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Response')
plt.grid()
plt.show()
```
